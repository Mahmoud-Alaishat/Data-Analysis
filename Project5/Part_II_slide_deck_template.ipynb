## Investigation Overview


> In this investigation, I wanted to Look at the gender of riders as well as the birth year and user type to see if they are related to the duration of trips


## Dataset Overview

>  This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area with about 183412 individuals.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
bike_trips = pd.read_csv('201902-fordgobike-tripdata.csv')
```

##   Genders in the data

>The distribution of genders and how they looks between males females and other



```python
# plot the gender for members
counts= bike_trips.member_gender.value_counts()
plt.figure(figsize=[12, 8])
colors = ['#66b3ff' ,'#ff9999','#99ff99']
plt.pie(counts,labels=counts.index,autopct='%1.0f%%',pctdistance=1.1, labeldistance=1.2,colors=colors);
plt.title('The distribution of genders');
```

## User types in the data

> The count of Users Types (subscriber ,customer)

```python
# plot the user types using countplot
counts= bike_trips.user_type.value_counts()
plt.figure(figsize=[8, 6])

sb.countplot(data=bike_trips,x='user_type',color=sb.color_palette()[0]);
plt.title('Types of Users')
plt.xlabel('User Type');

```

## Gender in years
>Gender according to the birth year which appears to be centralized between 1980 and 2000 for all genders



```python
plt.figure(figsize=[8, 6])

sb.violinplot(data=bike_trips,x='member_birth_year',y= 'member_gender',palette='Set2')
plt.xlabel('Birth Year')
plt.ylabel('Gender')
plt.title('The centralization of gender according to the birth year');
```

## User Type VS Duration of trips
>Does the type of user affect the duration of trips, IT turns out that the answer is yes take a look at the plot below

```python
# see the relation between Duration and Member User Type

plt.figure(figsize=[8, 5])
sb.barplot(data=bike_trips, x='user_type', y='duration_sec');
plt.xlabel('User Type')
plt.ylabel('Trips duration (sec) ')
plt.title('Duration of trips for User Customer and Subscriber');
```
